```python
# III. Determine the accuracy of the Coordinates with Address
geolocator = Nominatim(user_agent="myGeocoder")

def calculate_accuracy(row):
    location = geolocator.geocode(row['Address'])
    if location is not None:
        address_coords = (location.latitude, location.longitude)
        given_coords = (row['Latitude'], row['Longitude'])
        distance = geodesic(address_coords, given_coords).meters
        accuracy = max(0, 100 - (distance // 10))
        return accuracy
    return None

```
