```python
# Further classify the commercial vehicles
year_data['Subcategory'] = year_data['Vehicle Type'].apply(lambda x: 'LMV' if x in ['Bus', 'Truck', 'Trailer', 'Tractor', 'Tanker', 'Stage/Contract Carriage'] else 'HMV')
```

```python
# Plot the graph
year_data['Category'].value_counts().plot(kind='bar', color=['blue', 'orange', 'green'])
```
